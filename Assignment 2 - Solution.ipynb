## 1. Linear Algebra
### Goal: understand the basics of matrix algebra with a toy example

```python
import numpy as np
```

1.1 Generate a numpy array with 16 random normally distributed numbers.
Set the seed of random number generator to the five digits of your netid (removing leading 0)

```python
np.random.seed(16107)
v = np.random.randn(16)
v
```

1.2 Reshape the numpy array to a 4*4 matrix M

```python
M = v.reshape((4,4))
M
```

1.3 Compute the determinant and rank of the matrix

```python
np.linalg.det(M)
```

```python
np.linalg.matrix_rank(M)
```

1.4 Check if A=MM' and B=M'M are the same (difference no more than 1e-6 on each element)

```python
A = M.dot(M.T)
A
```

```python
B = M.T.dot(M)
B
```

```python
C = abs(A-B)<1e-6
C.all()
```

1.5 Compute the inverse of the matrix A=MM'

```python
Ai = np.linalg.inv(A)
Ai
```

1.6 Compute the product of A and its inverse (if exists)

```python
A.dot(Ai)
```

1.7 Set the diagnal elements of matrix M to 0. What are the determinant and rank of M now?

```python
np.fill_diagonal(M, 0)
M
```

```python
np.linalg.det(M)
```

```python
np.linalg.matrix_rank(M)
```

1.8 Multiple matrix M in 1.7 by a matrix X (on the left or right), such that every element in the i-th column is multiplied by scalar i (i=1,2,3,4)

```python
np.diag([1, 2, 3, 4])
```

```python
M.dot(np.diag([1, 2, 3, 4]))
```

### 2.2 Data Exploration

a. For each dataset, check the numbers of observations and missing values. Summarize the results in a easy to read format, such as a dictionary, matrix or data frame

```python
obs = [[len(df.index),np.count_nonzero(np.isnan(df.VALUE))]  for df in frames]
obs = pd.DataFrame(obs, index=keys, columns=['obs','missing'])
obs
```

b. Extract the observations between 2017-08-01 and 2017-08-31 for each dataset. <br>
You will work on the extracted subsets from now on

```python
for df in frames: df.DATE = pd.to_datetime(df.DATE)
```

```python
subframes = [df[df.DATE.isin(pd.date_range('2017-08-01','2017-08-31')) ] for df in frames]
subframes[6]
```

c. summarize the descriptive stats (pd.describe) of the datasets in a easy to read format, such as a dictionary, matrix or data frame

```python
pd.concat([df.describe() for df in subframes], axis=1, keys=keys)
```

### 2.3 Concatenation

a. Concat the extracted datasets, along the rows, to one big data frame. Add a column to indicate which Stock Market Index each observation is about

```python
df1 = pd.concat(subframes, ignore_index=True)
df1['SI'] = np.repeat(keys, [len(df.index) for df in subframes])
df1
```

b. The resulting dataset above is in a long format. Now convert it into a wide format, with row labels being dates and column labels being names of Stock Market Indexes

```python
df1.pivot(index='DATE',columns='SI',values='VALUE')
```

### 2.4 Merging Data

a. Merge the extracted datasets into one data frame. The resulting data frame should be in a wide format with one column labeled by Date and the rest columns labeled being names of Stock Market Indexes 

```python
subframes = [df.rename(columns={'VALUE':key}) for key, df in zip(keys, subframes)]
```

```python
df2 = subframes[0]
for i in range(1,len(subframes)): df2 = df2.merge(subframes[i], on='DATE')
df2
```

b. Convert the wide format dataset above into a long format

```python
df2.melt(id_vars='DATE')
```

c. Back to the wide format data frame, set the DATE column as index (row labels).

```python
df3 = df2.set_index('DATE')
df3
```

d. Add weekend dates to the data frame in step c, set the values on weekends as 0.

```python
df4 = df3.reindex(index=pd.date_range('2017-08-01','2017-08-31'), fill_value=0)
df4
```

e. Save the wide format dataset above as wide.csv, respectively. Submit the data on HusckCT as well
